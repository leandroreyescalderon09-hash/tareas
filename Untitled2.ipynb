```python
 def show_name(name):
    print("Hello", name)

show_name("Leandro")
```

```python
def add(a, b):
    return a + b

print(add(3, 5))
```

```python
def subtract(a, b):
    return a - b
print(subtract(5, 3))
```

```python
def multiply(a, b):
    return a * b
print(multiply(3, 5))
```

```python
def divide(a, b):
    return a / b
print(divide(10, 2))
```

```python
def is_even(number):
    return number % 2 == 0
print(is_even(4))
print(is_even(7))
```

```python
def find_max(a, b):
  if a > b:
        return a
  else:
        return b
print(find_max(9,6))
```

```python
def square(number):
    return number ** 2
print(square(4))
```

```python
def cube(number):
    return number ** 3
print(cube(3))
```

```python
def count_letters(word):
    return len(word)
print(count_letters("Leandro"))
```

```python
def convert_to_upper(text):
    return text.upper()
print(convert_to_upper("leandro"))
```

```python
def calculate_area(length, width):
    return length * width
print(calculate_area(4, 6))
```

```python
def average(numbers):
    return sum(numbers) / len(numbers)
print(average([1,7,6,10,8,12,17]))
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
print(factorial(5))
```
